```python
diaPath = 'Diamond1'
MCEPath = 'MCE'
MCEAtoms = 0

def diaExtract(path):
    edge = '////////////////////////////////////////////////////////////'
    file_in = path + '.pov'
    infile = open(file_in, 'r')
    
    file_out = path + '.inc'
    outfile = open(file_out, 'w')
    for line in infile:
        if line.strip() == edge:
#            print(line)
            break
    for line in infile:
        if 'rotate' in line.strip():
            print('  rotate <drx,dry,drz>\n', file=outfile)
            print('  scale <sx,sy,sz>\n', file=outfile)
            print('  translate <tx,ty,tz>\n', file=outfile)
            print('  rotate <rx,ry,rz>\n', file=outfile)

        if 'rotate' not in line.strip():
            print(line, end='', file=outfile)

    infile.close()
    outfile.close()
    
    
def MCEExtract(path,MCEatoms):
    edge = 'background { color rgb <0.000000, 0.000000,0.000000>}'
    end = 'pigment {color rgb <1.000000, 0.000000, 0.000000>}}}'
    file_in = path + '.pov'
    infile = open(file_in, 'r')
    
    file_out = path + '.inc'
    outfile = open(file_out, 'w')
    
    print('#declare bondsColor = color rgb <1,1,1>;\n', end='', file=outfile)
    for line in infile:
        if line.strip() == edge:
#            print(line)
            break
    for line in infile:
        if line.strip() == end:
            break
        if MCEAtoms == 0:
            if 'AtomRadius' not in line.strip():
                if 'bondcylRadius' not in line.strip():
                    print(line, end='', file=outfile)
        if MCEAtoms == 1:
            print(line, end='', file=outfile)
            
    print('}}\n', file=outfile)
    print('rotate <rx,ry,rz>\n', file=outfile)
    print('}', file=outfile)
    infile.close()
    outfile.close()    

def povSplit(diaPath,MCEPath):
    diaExtract(diaPath)
    MCEExtract(MCEPath,MCEAtoms)
    
    defaultPov = 'default.pov'
    infile = open(defaultPov, 'r')
 
    file_out = diaPath + '_spit.pov'
    outfile = open(file_out, 'w')
    
    for line in infile:      
        print(line, end='', file=outfile)
    
    incDia = '#include "' + diaPath + '.inc"'
    incMCE = '#include "' + MCEPath + '.inc"\n'    
    print(incDia, file=outfile) 
    print(incMCE, file=outfile) 
    
    infile.close()
    outfile.close()
 
        
povSplit(diaPath,MCEPath)        
```
